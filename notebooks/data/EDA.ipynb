```python
import pandas as pd

# Load the diamonds dataset
df = pd.read_csv('diamonds.csv')

# Check for duplicates
num_duplicates = df.duplicated().sum()
print(f'Number of duplicate records: {num_duplicates}')

# Drop duplicates
df.drop_duplicates(inplace=True)

# Save the modified dataset to a new file
df.to_csv('diamonds_modified.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_columns = ['cut', 'color', 'clarity']

plt.figure(figsize=(8,6))

for i in categorical_columns:
    g = sns.catplot(data=df, x=i, kind='count', height=4, aspect=1.5)
    g.set(title='Distribution of {}'.format(i), xlabel=i, ylabel='Count')
    plt.show()
```
